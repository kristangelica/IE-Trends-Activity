```python
### Dropping Rows
naia_df.drop(index=0)
```

```python
## Exploring the data
plt.figure(figsize=(16,6))
naia_df['TMEAN'].plot()
```

```python
"""
Cleaning Rules:
T - 0.01
"""

### Check Data Types
naia_df['RAINFALL'] = naia_df['RAINFALL'].replace({"T":0.01})
# naia_df['RAINFALL'] = naia_df['RAINFALL'].replace({"T":0.01})
naia_df
```

```python
### Converting Data Types Again!
naia_df = naia_df.astype(float)
```
